```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#데이터셋 로드하기
df = sns.load_dataset('flights')
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
# barplot
plt.figure(figsize=(15,4)) #사이즈
sns.barplot(data=df,x='year',y='passengers',ci='sd') #검은색은 신뢰구간
```

```python
# lineplot
plt.figure(figsize=(15,4)) #사이즈
sns.lineplot(data=df,x='year',y='passengers') #색깔은 신뢰구간
```

```python
# pointplot
plt.figure(figsize=(15,4)) #사이즈
sns.pointplot(data=df,x='year',y='passengers') #검은색은 신뢰구간
```

```python
# barplot에 hue를 통해 month 값에 따라 여러 색으로 나눠 그리기
plt.figure(figsize=(15,4)) #사이즈
sns.barplot(data=df,x='year',y='passengers',hue='month') 
```

```python
# lineplot에 hue를 통해 month 값에 따라 여러 색으로 나눠 그리기
plt.figure(figsize=(15,4)) #사이즈
sns.lineplot(data=df,x='year',y='passengers',hue='month') 
```

```python
# lineplot에 hue를 통해 month 값에 따라 여러 색으로 나눠 그리기
plt.figure(figsize=(15,4)) #사이즈
sns.lineplot(data=df,x='year',y='passengers',hue='month')
plt.legend(bbox_to_anchor=(1.02,1),loc=2,borderaxespad=0.)
```

```python
# lineplot에 hue를 통해 year 값에 따라 여러 색으로 나눠 그리기
plt.figure(figsize=(15,4)) #사이즈
sns.lineplot(data=df,x='month',y='passengers',hue='year')
plt.legend(bbox_to_anchor=(1.02,1),loc=2,borderaxespad=0.)
```

```python
# boxplot
plt.figure(figsize=(15,4))
sns.boxplot(data=df,x='year',y='passengers')
```

```python
# boxenplot
plt.figure(figsize=(15,4))
sns.boxenplot(data=df,x='year',y='passengers')
```

```python
# violinplot

plt.figure(figsize=(15,4))
sns.violinplot(data=df,x='year',y='passengers')
```

```python
# 산점도

plt.figure(figsize=(15,4))
sns.swarmplot(data=df,x='year',y='passengers')
```

```python
# boxplot

plt.figure(figsize=(15,4))
sns.boxplot(data=df,x='month',y='passengers')
```

```python
# violinplot
plt.figure(figsize=(15,4))
sns.violinplot(data=df,x='month',y='passengers')
```

```python
# regplot

plt.figure(figsize=(15,4))
sns.regplot(data=df,x='year',y='passengers')
```

pivot으로 데이터 형태 바꾸기

```python
df
```

```python
df_h = df.pivot(index='month',columns='year',values='passengers')
df_h
```

```python
# heatmap
plt.figure(figsize=(15,5))
sns.heatmap(df_h,annot=True,fmt='.0f',cmap='Blues') #cmap:색깔
```

```python
# distplot
sns.distplot(df['passengers'])
```

서브플롯 그리기

```python
# pairplot
sns.pairplot(data=df, hue='month') #월별로 색을 다르게 한다.
```

```python
# relplot

sns.relplot(data=df, x='month', y='passengers')
```

```python
# relplot

sns.relplot(data=df, x='month', y='passengers', kind='line', hue='year',col='year')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats # 폰트 선명
set_matplotlib_formats('retina')
```

```python
url = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"
url
```

```python
table = pd.read_html(url)
len(table)
```

```python
table[1]
```

```python
df = table[2]
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.set(font='Malgun Gothic') #윈도우 폰트
plt.figure(figsize=(15,4)) #Shift+tab 키로 옵션볼수있음
plt.xticks(rotation=60) #글씨가 기울어짐 
sns.pointplot(data=df, x='연도 (년)',y='추계인구(명)')
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df, x='연도 (년)',y='추계인구(명)') #구간별로 나눔
```

```python
plt.figure(figsize=(15,4)) #굵게 표시
plt.xticks(rotation=60) #글씨가 기울어짐
sns.pointplot(data=df, x='연도 (년)',y='출생자수(명)')
```
